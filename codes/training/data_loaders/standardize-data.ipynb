```python
import numpy as np 
import pandas as pd 
import os
import glob
import cv2
import random
import matplotlib.pyplot as plt
```

```python
x_train = np.load("x_train_200.npy")
y_train = np.load("y_train_200.npy")
x_train.shape, y_train.shape
```

```python
x_val = np.load("x_val_200.npy")
y_val = np.load("y_val_200.npy")
x_val.shape, y_val.shape
```

```python
x_test = np.load("x_test_200.npy")
y_test = np.load("y_test_200.npy")
x_test.shape, y_test.shape
```

```python
def n2(img):
    mean = np.mean(img,axis=(0,1,2))
    std = np.std(img,axis=(0,1,2))
    img_norm = (img-mean)/(std+1e-7)
    return img_norm
```

```python
x_train_std = []

for img in x_train[:100]:
    img_std = None
    img_std = n2(img)
    x_train_std.append(img_std)

x_train_std = np.array(x_train_std)
x_train_std.shape
```

```python
plt.imshow(x_train_std[50])
```

```python
def all_norm(images):
    images_std = []
    for img in images:
        img_std = None
        img_std = n2(img)
        images_std.append(img_std)

    images_std = np.array(images_std)
    return images_std
    
```

```python
x_train_std = all_norm(x_train)
x_train_std.shape
```

```python
x_test_std = all_norm(x_test)
x_test_std.shape
```
